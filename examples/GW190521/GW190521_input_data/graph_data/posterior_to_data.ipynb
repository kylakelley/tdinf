```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

f = h5py.File("../H-H1_GWOSC_16KHZ_R2-1242442952-32.hdf5", "r")

strain = f["strain"]["Strain"][:]
# Time spacing of LIGO data is 61 µs
dt = 0.000061
times = dt * np.arange(len(strain))

plt.plot(times, strain)
plt.xlabel('Time [s]')
plt.ylabel('Strain')
plt.show()
```

```python
samples = group["posterior_samples"]

print(type(samples))
print(samples.shape)
print(samples.dtype)
```

```python
samples.dtype.names
```

```python
total_mass_post = samples["total_mass"][:]
mass_ratio_post = samples["mass_ratio"][:]
luminosity_distance_post = samples["luminosity_distance"][:]
inclination_post = samples["iota"][:]

s1x = samples["spin_1x"][:]
s1y = samples["spin_1y"][:]
s1z = samples["spin_1z"][:]
s2x = samples["spin_2x"][:]
s2y = samples["spin_2y"][:]
s2z = samples["spin_2z"][:]

spin1_magnitude_post = np.sqrt(s1x**2 + s1y**2 + s1z**2)
spin2_magnitude_post = np.sqrt(s2x**2 + s2y**2 + s2z**2)
```

```python
plt.hist(total_mass, bins=100, density=True, label="output")
plt.hist(total_mass_post, bins=100, density=True, label="posterior")
plt.xlabel("Total Mass in solar massses")
plt.ylabel("Density")
plt.show()
```

```python
rng = np.random.default_rng()
M = rng.uniform(0, 100, size=10000)
```

```python
print(M)
```

```python
rng = np.random.default_rng()
q = rng.uniform(0, 1, size=10000)
```

```python
print(q)
```

```python
m1 = M/(1+q)
m2 = M-m1
```

```python
print(m1, m2)
```

```python
plt.scatter(m1, m2, s=1)
plt.show()
```

```python
example_weights = 1/(M/m1**2)/100

inds_to_keep = np.where(np.random.random(10000)<example_weights)

plt.scatter(m1[inds_to_keep], m2[inds_to_keep], s=10)
```

```python
mass1_post = total_mass_post/(1+mass_ratio_post)
weights = total_mass_post/(mass1_post)**2
```

```python
plt.hist(total_mass, bins=100, density=True, label="output")
plt.hist(total_mass_post, bins=100, density=True, label="posterior", weights=weights, alpha=0.8)
plt.xlabel("Total Mass in solar massses")
plt.ylabel("Density")
plt.show()
```

```python
plt.hist(mass_ratio, bins=40, density=True, label="output")
plt.hist(mass_ratio_post, bins=40, density=True, label="posterior", weights=weights, alpha=0.8)
plt.xlabel("Mass Ratio")
plt.ylabel("Density")
plt.show()
```
