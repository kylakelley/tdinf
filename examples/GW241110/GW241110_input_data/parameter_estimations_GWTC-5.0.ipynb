```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
f = h5py.File("GW241110_124123-combined_PEDataRelease")
print(list(f.keys()))
```

```python
list(f['C00:IMRPhenomXPNR'].keys())
```

```python
group = f['C00:IMRPhenomXPHM-SpinTaylor']
group2 = f['C00:IMRPhenomXPNR']
group3 = f['C00:SEOBNRv5PHM']
```

```python
samples = group['posterior_samples']
samples2 = group2['posterior_samples']
samples3 = group3['posterior_samples']
```

```python
total_mass_post = samples["total_mass_source"][:]
mass_ratio_post = samples["mass_ratio"][:]
spin_magnitude_1_post = samples["a_1"][:]
spin_magnitude_2_post = samples["a_2"][:]
luminosity_distance_post = samples["luminosity_distance"]
chi_eff_post = samples["chi_eff"]

total_mass_post2 = samples2["total_mass_source"][:]
mass_ratio_post2 = samples2["mass_ratio"][:]
spin_magnitude_1_post2 = samples2["a_1"][:]
spin_magnitude_2_post2 = samples2["a_2"][:]
luminosity_distance_post2 = samples2["luminosity_distance"]
chi_eff_post2 = samples2["chi_eff"]

redshift = samples["redshift"]
redshift2 = samples2["redshift"]
redshift3 = samples3["redshift"]

total_mass_post3 = samples3["total_mass_source"][:]
mass_ratio_post3 = samples3["mass_ratio"][:]
spin_magnitude_1_post3 = samples3["a_1"][:]
spin_magnitude_2_post3 = samples3["a_2"][:]
luminosity_distance_post3 = samples3["luminosity_distance"]
chi_eff_post3 = samples3["chi_eff"]
```

```python
plt.hist(total_mass_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(total_mass_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(total_mass_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel(r"Total Mass ($M_{\odot}$)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
np.median(total_mass_post2)
```

```python
plt.hist(total_mass_post*(1+redshift), bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(total_mass_post2*(1+redshift2), bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(total_mass_post3*(1+redshift3), bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel(r"Total Mass ($M_{\odot}$)")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.hist(mass_ratio_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(mass_ratio_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(mass_ratio_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel("Mass Ratio")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.hist(spin_magnitude_1_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(spin_magnitude_1_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(spin_magnitude_1_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel("Spin Magnitude 1")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.hist(spin_magnitude_2_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(spin_magnitude_2_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(spin_magnitude_2_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel("Spin Magnitude 2")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.hist(luminosity_distance_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(luminosity_distance_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(luminosity_distance_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel("Luminosity Distance")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.hist(chi_eff_post, bins=100, density=True, alpha=.5, label="IMRPhenomXPHM-SpinTaylor")
plt.hist(chi_eff_post2, bins=100, density=True, alpha=.5, label="IMRPhenomXPNR")
plt.hist(chi_eff_post3, bins=100, density=True, alpha=.5, label="SEOBNRv5PHM")
plt.xlabel("Chi Effective")
plt.ylabel("Density")
plt.legend()
plt.show()
```
