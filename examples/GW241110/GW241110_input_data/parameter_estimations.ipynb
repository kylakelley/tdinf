```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = h5py.File("H-H1_GWOSC_O4b_16KHZ_R1-1415274496-4096.hdf5", "r")
l = h5py.File("L-L1_GWOSC_O4b_16KHZ_R1-1415274496-4096.hdf5", "r")
```

```python
h.visit(print)
```

```python
list(h.keys())
```

```python
h_strain = h["strain"]["Strain"][:]
h_GPSstart = h["meta"]["GPSstart"][()]
h_duration = h["meta"]["Duration"][()]
l_strain = l["strain"]["Strain"][:]
l_GPSStart = l["meta"]["GPSstart"][()]
l_duration = l["meta"]["Duration"][()]
```

```python
h_samples = len(h_strain)
l_samples = len(l_strain)
print(h_samples, l_samples)
```

```python
sample_rate_h = len(h_strain)/h_duration
sample_rate_l = len(l_strain)/l_duration
print(sample_rate_h, sample_rate_l)
```

```python
dt = 1/16384
time = np.arange(0, h_duration, dt)
```

```python
plt.plot(time, 0.5*h_strain, alpha=0.5, label="H")
plt.plot(time, l_strain, alpha=0.5, label="L")
plt.xlabel="Time"
plt.ylabel="Strain"
plt.show()
```
