```python
import numpy as np
import h5py
import matplotlib.pyplot as plt

f = h5py.File('PEDataRelease_mixed_cosmo.h5')
```

```python
f.keys()
```

```python
f['C01:IMRPhenomXPHM'].keys()
```

```python
samples = f['C01:IMRPhenomXPHM']['posterior_samples']
samples2 = f['C01:Mixed']['posterior_samples']
samples3 = f['C01:SEOBNRv4PHM']['posterior_samples']
```

```python
samples.dtype.names
```

```python
total_mass = samples['total_mass']
mass_ratio = samples['mass_ratio']
luminosity_distance = samples ['luminosity_distance']
spin_1x = samples['spin_1x']
spin_1y = samples['spin_1y']
spin_1z = samples['spin_1z']
spin_2x = samples['spin_2x']
spin_2y = samples['spin_2y']
spin_2z = samples['spin_2z']

spin_1magnitude = np.sqrt(spin_1x **2 + spin_1y **2 + spin_1z **2)
spin_2magnitude = np.sqrt(spin_2x **2 + spin_2y **2 + spin_2z **2)
```

```python
plt.hist(total_mass, bins=100, density=True)
plt.xlabel('Total Mass')
plt.ylabel('Density')
```

```python
plt.hist(mass_ratio, bins=100, density=True)
plt.xlabel('Mass Ratio')
plt.ylabel('Density')
```

```python
plt.hist(spin_1magnitude, bins=100, density=True)
plt.xlabel('Spin 1 Magnitude')
plt.ylabel('Density')
```

```python
plt.hist(spin_2magnitude, bins=100, density=True)
plt.xlabel('Spin 2 Magnitude')
plt.ylabel('Density')
```

```python
plt.hist(luminosity_distance, bins=100, density=True)
plt.xlabel('Luminosity Distance')
plt.ylabel('Density')
```

```python
total_mass2 = samples2['total_mass']
mass_ratio2 = samples2['mass_ratio']
luminosity_distance2 = samples2['luminosity_distance']
spin_1x = samples2['spin_1x']
spin_1y = samples2['spin_1y']
spin_1z = samples2['spin_1z']
spin_2x = samples2['spin_2x']
spin_2y = samples2['spin_2y']
spin_2z = samples2['spin_2z']

spin_1magnitude2 = np.sqrt(spin_1x **2 + spin_1y **2 + spin_1z **2)
spin_2magnitude2 = np.sqrt(spin_2x **2 + spin_2y **2 + spin_2z **2)
```

```python
total_mass3 = samples3['total_mass']
mass_ratio3 = samples3['mass_ratio']
luminosity_distance3 = samples3['luminosity_distance']
spin_1x = samples3['spin_1x']
spin_1y = samples3['spin_1y']
spin_1z = samples3['spin_1z']
spin_2x = samples3['spin_2x']
spin_2y = samples3['spin_2y']
spin_2z = samples3['spin_2z']

spin_1magnitude3 = np.sqrt(spin_1x **2 + spin_1y **2 + spin_1z **2)
spin_2magnitude3 = np.sqrt(spin_2x **2 + spin_2y **2 + spin_2z **2)
```

```python
plt.hist(total_mass, bins=100, density=True, alpha=0.5, label= 'IMRPhenomXPHM')
plt.hist(total_mass2, bins=100, density=True, alpha=0.5, label= 'Mixed')
plt.hist(total_mass3, bins=100, density=True, alpha=0.5, label= 'SEOBNRv4PHM')
plt.legend()
plt.xlabel('Total Mass')
plt.ylabel('Density')
plt.show()
```
